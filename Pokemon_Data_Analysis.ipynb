##Visualization

```python
# Create boxplot to find the relation between attack and special attack
sns.boxplot(data=df, x='Attack', y='Sp. Atk')
```

```python
# Create scatterplot to find the relation between Attack and Sp. Atk
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Attack', y='Sp. Atk', alpha=0.5, color='red')
plt.title('Attack vs Sp. Atk')
plt.xlabel('Attack')
plt.ylabel('Sp. Atk')
plt.grid(True)

plt.show()
```

###Check Distribution Data

```python
# Create boxplot to observe distribution of numerical value
import seaborn as sns

fig, axs = plt.subplots(1,2, figsize=(15,5))
sns.boxplot(df['Attack'], ax = axs[0])
sns.boxplot(df['HP'], ax = axs[1])
```

```python
# Remove Outlier
# Attack
qg25, qg50, qg75 = np.percentile(df['Attack'], [25, 50, 75])
iqrg = qg75 - qg25
ming = qg25 - 1.5 * iqrg
maxg = qg75 + 1.5 * iqrg
df = df[(df['Attack'] < maxg)]

# HP
qh25, qh50, qh75 = np.percentile(df['HP'], [25, 50, 75])
iqrh = qh75 - qh25
lower_bound = qh25 - 1.5 * iqrh
upper_bound = qh75 + 1.5 * iqrh
df = df[(df['HP'] >= lower_bound) & (df['HP'] <= upper_bound)]
```

```python
# Removing Outliers
fig, axs = plt.subplots(1,2, figsize=(15,5))
sns.histplot(data = df, x = 'Attack', kde = True, color = 'skyblue', ax = axs[0])
sns.histplot(data = df, x = 'HP', kde = True, color = 'olive', ax = axs[1])

plt.show()
```

###Count

```python
# Defining the labels & colors
labels = ['No Evolution', 'Evolution']
pastel = ['pink', 'skyblue']

# Counting the number of Pokemon with and without evolution (Type 2)
evolution_counts = df['Type 2'].isnull().value_counts()

# Plotting the pie chart
plt.figure(figsize=(7,7))
plt.pie(evolution_counts, labels=labels, colors=pastel, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Pokemon with and without Evolution')
plt.axis('equal')

plt.show()
```

```python
# Group the Pokemon data based on Generation and find total Pokemon in each Generation
total_per_generation = df.groupby('Generation')['#'].count()

print("Total Pokemon in each generation:")
print(total_per_generation)
```

```python
# Defining the colors of the bar chart
colors = ['#DD4F4A', '#F1B652', '#FFFD4B', '#B6F694', '#BDE8F1', '#D19ADF']

plt.bar(total_per_generation.index, total_per_generation.values, color=colors)

# Add title
plt.xlabel('Generation')
plt.ylabel('Total Pokemon')
plt.title('Total Pokemon in Each Generation')

# Show the bar chart
plt.show()
```

```python
# Changing "True to "Legendary" and False to "Non Legendary"
df['Legendary'] = df['Legendary'].replace({True: 'Legendary', False: 'Non Legendary'})

# Number of Legendary Pokemon
number_of_pokemons = df['Legendary'].value_counts()

print(number_of_pokemons)
```

```python
# Defining the labels & colors
labels = ['Legendary', 'Non Legendary']
colors = ['gold', 'olive']

# Counting the number of Legendary and Non Legendary Pokemon
df['Legendary'] = df['Legendary'].replace({True: 'Legendary', False: 'Non Legendary'})
Legendary_counts = df['Legendary'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(7,7))
plt.pie(Legendary_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=0)
plt.title('Percentage of Legendary Pokemon')
plt.axis('equal')

plt.show()
```
